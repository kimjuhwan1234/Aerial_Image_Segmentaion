```python
import os

def rename_files_with_folder_name(folder_path):
    for root, _, files in os.walk(folder_path):
        folder_name = os.path.basename(root)
        for filename in files:
            if filename.endswith('.jpg') or filename.endswith('.png') or filename.endswith('.json'):
                new_filename = f"1102_{folder_name}_{filename}"
                file_path = os.path.join(root, filename)
                new_file_path = os.path.join(root, new_filename)
                os.rename(file_path, new_file_path)
                print(f"Renamed: {filename} -> {new_filename}")

folder_path = "/mnt/disks/data1/radar/레이더 데이터old_radar/old_radar/extract_frame_1102/extract_frame8"
rename_files_with_folder_name(folder_path)
```

```python
import os
import shutil

def merge_folders(source_folder, destination_folder):
    for root, _, files in os.walk(source_folder):
        for filename in files:
            if filename.endswith('.jpg') or filename.endswith('.png') or filename.endswith('.json'):
                source_path = os.path.join(root, filename)
                destination_path = os.path.join(destination_folder, filename)
                shutil.copy2(source_path, destination_path)  # 변경된 부분
                print(f"Copied: {filename}")

source_folder = "/mnt/disks/data1/radar/레이더 데이터old_radar/old_radar/extract_frame_1102/extract_frame8"
destination_folder = "/mnt/disks/data1/radar/레이더 데이터old_radar/old_radar/extract_frame_1102/integrated"

if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

merge_folders(source_folder, destination_folder)
```
